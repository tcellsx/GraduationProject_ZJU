```python
from selenium import webdriver
```

### ChromeDriver记得放到当前路径（系统环境变量也行）

```python
driver = webdriver.Chrome()
```

```python
url = "https://www.xinpianchang.com/discover/article?from=navigator"
```

### 打开页面的API

```python
driver.get(url)
```

#### 再次打开一个页面

```python
url2 = "https://www.baidu.com"
```

```python
driver.get(url2)
```

#### 回到上一个页面

```python
driver.back()
```

#### 前进到下一个页面

```python
driver.forward()
```

### 与页面当中的元素进行交互
#### 1.找到元素
#### 2.声明操作

```python
element = driver.find_element("id","kw")
```

### 向输入框中发送值

```python
element.send_keys("爬虫")
```

### 寻找按钮

```python
element2 = driver.find_element("id","su")
```

### 点击按钮

```python
element2.click()
```

### 获取页面当中的内容 

```python
driver.page_source
```

### 关闭页面

```python
driver.close()
```
