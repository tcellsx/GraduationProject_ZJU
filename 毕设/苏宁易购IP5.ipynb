```python
import requests
import re
from time import sleep
import json

texts1 = []
texts2 = []
texts3 = []
texts4 = []

# 定义函数来获取评论数据并处理
def get_comments(page):
    url_comment = 'https://review.suning.com/ajax/cluster_review_lists/cluster-38346611-000000012426366980-0000000000-total-'+str(page)+'-default-10-----reviewList.htm?callback=reviewList'
    headers = {
        'referer': 'https://product.suning.com/',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36',
        'Cookie' : '_snmc=1; _snsr=baidu.com%7Corganic%7C%7C%7C*%3A*; _snvd=1711701853406fsHUNd5iUjb; streetCode=5730199; cityCode=573; districtId=12517; cityId=9317; _snzwt=THtOlO18e8962a0659a8E1c1e; authId=siKHyp6Qn4F0V1PadTgevboAZCX6UxVzUu; secureToken=F1F7CEBA47D50077A98838DE70F95EC7; ssotbrd=TGTpFMOvajtX4V06PkwdTo8Y8cRymAg11W1fcUvxiio; tradeMA=114; hm_guid=2d091b02-4446-43bc-92f7-fe9292e2c5c0; _df_ud=5caa4aa0-1e75-4d20-a7e0-f2e1acee6dea; totalProdQty=0; _snms=171170193562342705; _snadtp=1; _snadid=HT_40196840_888888888_12426366980_20240329; smhst=12426366980|0000000000; token=1dc60bbe-37fb-4212-b887-f402f643be7a; SN_CITY=130_573_1000325_9317_01_12517_1_1_99_5730199; _snma=1%7C171170192516423700%7C1711701925164%7C1711701935616%7C1711701960985%7C3%7C1; _snmp=171170196045960954; _snmb=171170192529522816%7C1711701960991%7C1711701960986%7C3; route=e67b78cb212da195818d6cebc7f3ff42'  # 请替换为您的Cookie
    }
    sleep(3)
    response = requests.get(url_comment, headers=headers)
    print(url_comment)

    # 使用正则表达式提取 JSON 数据部分  jsonp
    json_data_match = re.search(r'reviewList\((.*)\)', response.text,re.S)
    print(response.text)
    if json_data_match:
        json_data = json_data_match.group(1)
        # 解析 JSON 数据
        data = json.loads(json_data)
        print(data)
        
        # 处理评论数据
        for item in data.get('commodityReviews', []):
            texts1.append(item.get('content', ''))
            texts2.append(item.get('publishTime', ''))
            texts3.append(str(item.get('qualityStar', '')))  # 分数转换为字符串
            texts4.append(item.get('ipLocationName', ''))
        
        # 在获取评论数据后添加以下代码，用于将数据存储为 txt 文件
        with open('comments_data.txt', 'w', encoding='utf-8') as f:
            for content, publish_time, quality_star, location in zip(texts1, texts2, texts3, texts4):
                f.write(f"评论内容: {content}\n")
                f.write(f"评论时间: {publish_time}\n")
                f.write(f"评分: {quality_star}\n")
                f.write(f"位置: {location}\n\n")
        with open('comments_data_content.txt', 'w', encoding='utf-8') as f1:
            for content in texts1:
                f1.write(f"评论内容: {content}\n\n")

        with open('comments_data_creatime.txt', 'w', encoding='utf-8') as f2:
            for publish_time in texts2:
                f2.write(f"评论时间: {publish_time}\n")

        with open('comments_data_score.txt', 'w', encoding='utf-8') as f3:
            for quality_star in texts3:
                f3.write(f"评分: {quality_star}\n")

        with open('comments_data_location.txt', 'w', encoding='utf-8') as f4:
            for location in texts4:
                f4.write(f"位置: {location}\n")

    else:
        print("No JSON data found in response.")

# 循环获取评论数据
for page in range(1,200):  # 这里只爬取一页
    get_comments(page)
    print('爬完了第' + str(page + 1) + '页')
```
